```python
#0.5 Wir sind a) an der Gesamtanzahl der Director- und Actor_{1,2,3}-Likes interessiert.
#Um nicht jedes Mal die Summe über alle bilden zu müssen, empfiehlt sich mit „direc-
#tor_actor_facebook_likes“ eine weitere Spalte, die die Summe der definierten Likes je Film
#aufweist. Bilden Sie diese.

cols = ["director_facebook_likes", "actor_1_facebook_likes", "actor_2_facebook_likes", "actor_3_facebook_likes"]
movie_df["director_actor_facebook_likes"] = movie_df.apply(lambda row: row[cols].sum(),axis=1)
# oder
# movie_df["director_actor_facebook_likes"] = movie_df["actor_1_facebook_likes"] + movie_df["actor_2_facebook_likes"] + movie_df["actor_3_facebook_likes"] + movie_df["director_facebook_likes"]
```

```python
#b) Wie man der exemplarischen Ergebnisdarstellung entnehmen kann, weist zumindest ein
#Eintrag den Wert „NaN“ auf. Was ist die Ursache?

# Wenn eine Zeile einen NaN Wert enthält, wird das Ergebnis automatisch NaN, egal wenn nur ein Wert NaN aufweist und die anderen einen Wert haben
# NaN + 5 = NaN
```

```python
#Finden Sie c) heraus, um wie viele Einträge es tatsächlich handelt und eliminieren Sie d) das
#Problem, indem Sie die betroffenden Felder des Datasets auf „0“ setzen.

# Summe der Null-Werte
movie_df["director_actor_facebook_likes"].isna().sum()
# Nan Einträge durch 0 erstetzen
movie_df["director_actor_facebook_likes"] = movie_df["director_actor_facebook_likes"].fillna(0)
```
