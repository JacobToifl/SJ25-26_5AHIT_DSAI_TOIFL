```python
import pandas as pd
from sympy.integrals.meijerint_doc import category
#import necessary libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D, Dense
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import keras
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from keras.utils import load_img
import zipfile
```

```python
# entpackt alle Ordner und Dateien im zip-Archiv in das aktuelle Verzeichnis
zip_path = "cat_dog-train.zip"

with zipfile.ZipFile(zip_path, "r") as zip:
    zip.extractall(".")
```

```python
# danach kann ich einfach den Ordner "train" durchsuchen, weil zuvor entpackt wurde
train = []

for file_name in os.listdir("train"):
    if file_name.endswith(".jpg"):
        category = file_name.split(".")[0]   # cat, dog

        train.append({
            "filename": file_name,      # nur der Name
            "category": category
        })

train_df = pd.DataFrame(train)
train_df.head()

# dieser Code liest einfach nur die Daten aus dem zip-Archiv ein, ohne sie zu entpacken - kann dann
# problematisch werden, zum Beispiel bei der show_image() Funktion

#zip_path = "cat_dog-train.zip"

#train = []

#with zipfile.ZipFile(zip_path, "r") as zip_ref:
#    for file_name in zip_ref.namelist():
#        filename_only = file_name.split("/")[-1]
#        category = filename_only.split(".")[0]
#
#        train.append({
#            "filename": filename_only,
#            "category": category
#        })
#
#df = pd.DataFrame(train)
#df.head()
```

```python
train_dir = Path("train")

def show_img(filename):
    img_path = train_dir / filename   # train/cat.0.jpg
    img = load_img(img_path)
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
# Aufteilen der Trainingsdaten in train und valdation - hilft gegen Overfitting
train_data, val_data = train_test_split(train_df, test_size=0.2)
train_data = train_data.reset_index(drop=True)
val_data = val_data.reset_index(drop=True)
```

```python
# Shape der Daten ausgeben
print(train_data.shape)
print(val_data.shape)
```

```python
model = Sequential()

model.add(Conv2D(16, (3, 3), activation='relu',padding='same', input_shape=(150, 150, 3)))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2)))
model.add(GlobalAveragePooling2D())

model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)
validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_dataframe(
    dataframe = train_data,
    directory = "train",
    x_col = "filename",
    y_col = "category",
    target_size = (150,150),
    batch_size = 64,
    class_mode = "binary"
)

validation_generator = validation_datagen.flow_from_dataframe(
    dataframe = val_data,
    directory = "train",
    x_col = "filename",
    y_col = "category",
    target_size = (150,150),
    batch_size = 64,
    class_mode = "binary"
)
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch = 312,
    validation_data = validation_generator,
    epochs=3
)
```
