```python
from pathlib import Path
import pandas as pd
from keras.utils import load_img
import matplotlib.pyplot as plt
```

```python
train_dir = Path("train")

train = []
for img_path in train_dir.iterdir():
    filename = img_path.name
    category = filename.split(".")[0]

    train.append({
        "filename": filename,
        "category": category
    })
```

```python
train = pd.DataFrame(train)
print(train.shape)
train.head()
```

```python
value_counts = train["category"].value_counts()

print(train.shape[0])
print(value_counts)
```

```python
train_dir = Path("train")

def show_img(filename):
    img_path = train_dir / filename
    img = load_img(img_path)
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
img_name = "cat.1000.jpg"
show_img(img_name)
```
